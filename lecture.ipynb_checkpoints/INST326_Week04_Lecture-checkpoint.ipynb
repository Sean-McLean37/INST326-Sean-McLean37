# ðŸ§  The Big Idea: From Functions to Objects
## (25 minutes)

### The Evolution of Your Programming

**Previous Weeks:**
- Week 1-3: Functions process data
- Week 4: Objects combine data + behavior

### Real-world Analogy
> **"A function is like a tool - it does one job. A class is like a complete workshop - it has tools (methods) and storage (attributes) all in one place."**

### Why Objects Matter
- **Better organization:** Related data and functions stay together
- **Easier to understand:** Code models real-world things
- **Reusable:** Create many objects from one class template
- **Scalable:** Build complex systems from simple objects

## Class Definition Basics

Let's start with a simple PlantingContainer class:

````python
class PlantingContainer:
    """Represents a container for planting in garden management."""

    def __init__(self, container_type, length, width, depth=12):
        """Initialize a new planting container."""
        self.container_type = container_type
        self.length = length
        self.width = width
        self.depth = depth
        self.location = None
        self.plants = []

    def calculate_area(self):
        """Calculate the planting surface area."""
        return self.length * self.width

    def calculate_volume(self):
        """Calculate soil volume capacity."""
        return self.length * self.width * self.depth

print("PlantingContainer class defined!")
````

### Key Teaching Points:
- `class` keyword creates a template
- `__init__` is the constructor - runs when you create an object
- `self` refers to the specific object instance
- Attributes store data, methods provide behavior

## Creating and Using Objects

Now let's create some objects and see them in action:

````python
# Creating objects (instances)
raised_bed = PlantingContainer("raised_bed", 48, 24, 8)
container_pot = PlantingContainer("pot", 12, 12, 10)

print("Objects created successfully!")
print(f"Raised bed type: {raised_bed.container_type}")
print(f"Container pot type: {container_pot.container_type}")
````

````python
# Using object methods
bed_area = raised_bed.calculate_area()
pot_volume = container_pot.calculate_volume()

print(f"Raised bed area: {bed_area} square inches")
print(f"Container pot volume: {pot_volume} cubic inches")
````

````python
# Objects maintain their own state
raised_bed.location = "backyard_south"
container_pot.location = "front_porch"

print(f"Bed location: {raised_bed.location}")
print(f"Pot location: {container_pot.location}")

# Each object is independent!
print(f"Bed dimensions: {raised_bed.length} x {raised_bed.width}")
print(f"Pot dimensions: {container_pot.length} x {container_pot.width}")
````

### Key Teaching Points:
- Each object is independent
- Objects remember their own data
- Method calls use dot notation
- Same class, different objects, different data

# ðŸ“‹ Requirements-Driven Class Design
## (20 minutes)

## From Requirements to Class Structure

Let's look at the requirements document:

> *"Planting containers include garden beds, pots, planters, bags, trays, etc. Each object should include the following attributes:*
> - *Container id*
> - *Container type (bed, raised bed, pot, planter, bag, tray, etc.)*
> - *Shape attributes - what shape is it?*
> - *Dimensional attributes - length, width, depth, radius, diameter*
> - *Location and mobility*
> - *Purpose (vegetables, herbs, flowers, fruit, other)"*

### Translation Process:

1. **Identify the class:** PlantingContainer (typically nouns)
2. **List attributes from requirements:** (typically characteristics or properties)
   - container_id, container_type, length, width, depth, location, mobility, purpose
3. **Determine methods from requirements:** (typically verbs)
   - "Calculate container area" â†’ `calculate_area()`
   - "Calculate container volume" â†’ `calculate_volume()`

## Enhanced Class Implementation

Let's build a more complete class based on requirements:

````python
class PlantingContainer:
    """Represents a container for planting in garden management."""

    def __init__(self, container_id, container_type, length, width, depth=12, purpose="mixed"):
        """Initialize a new planting container."""
        self.container_id = container_id
        self.container_type = container_type
        self.length = length
        self.width = width
        self.depth = depth
        self.purpose = purpose
        self.location = None
        # Smart default: beds are fixed, pots are movable
        self.mobility = "fixed" if container_type in ["bed", "raised_bed"] else "movable"
        self.plants = []

    def calculate_area(self):
        """Calculate the planting surface area in square inches."""
        return self.length * self.width

    def calculate_volume(self):
        """Calculate soil volume capacity in cubic inches."""
        return self.length * self.width * self.depth

    def get_planting_capacity(self, plant_spacing):
        """Calculate how many plants can fit with given spacing."""
        if plant_spacing <= 0:
            return 0

        plants_per_row = int(self.length // plant_spacing)
        rows = int(self.width // plant_spacing)
        return plants_per_row * rows

    def __str__(self):
        """String representation for easy display."""
        return f"{self.container_type} (ID: {self.container_id}): {self.length}x{self.width}x{self.depth} inches"

print("Enhanced PlantingContainer class defined!")
````

Let's test our enhanced class:

````python
# Create containers from requirements
bed1 = PlantingContainer("BED001", "raised_bed", 48, 24, 8, "vegetables")
pot1 = PlantingContainer("POT001", "pot", 12, 12, 10, "herbs")

# Show automatic mobility setting
print(f"Bed mobility: {bed1.mobility}")    # "fixed"
print(f"Pot mobility: {pot1.mobility}")    # "movable"

# Test the string representation
print(f"Bed info: {bed1}")
print(f"Pot info: {pot1}")
````

````python
# Calculate planting capacity
lettuce_capacity = bed1.get_planting_capacity(6)  # 6" spacing for lettuce
herb_capacity = pot1.get_planting_capacity(4)     # 4" spacing for herbs

print(f"Bed can hold {lettuce_capacity} lettuce plants (6\" spacing)")
print(f"Pot can hold {herb_capacity} herb plants (4\" spacing)")

# Show area and volume calculations
print(f"Bed area: {bed1.calculate_area()} sq inches")
print(f"Pot volume: {pot1.calculate_volume()} cubic inches")
````

# ðŸ”„ Transition: Functional to Object-Oriented Thinking
## (5 minutes)

## The Mental Shift

### Functional Thinking (Procedural Programming):
- "What functions do I need to solve this problem?"
- Functions operate on data passed as parameters
- Data and operations are separate

```python
# Functional approach
def calculate_area(length, width):
    return length * width

bed_length = 48
bed_width = 24
area = calculate_area(bed_length, bed_width)
```

### Object-Oriented Thinking:
- "What real-world things am I modeling?"
- Objects contain both data and the operations that make sense for that data
- Data and behavior are bundled together

```python
# Object-oriented approach
bed = PlantingContainer("BED001", "raised_bed", 48, 24)
area = bed.calculate_area()  # The bed knows how to calculate its own area!
```
