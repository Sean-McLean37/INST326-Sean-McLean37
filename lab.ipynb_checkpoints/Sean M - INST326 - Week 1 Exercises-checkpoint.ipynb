## Exercise 15: Call Number Parsing (No Lists)

Given `call_no = "QA.76.P98"`, extract `prefix`, `section`, and `cutter` using `.find()`/`.rfind()` and string slicing only (no `.split()`). Print them in one line.

```python
call_no = "QA.76.P98"
prefix = call_no[0:call_no.find('.')]
section = call_no[call_no.find('.') + 1:call_no.rfind('.')]
cutter = call_no[call_no.rfind('.') + 1:]

print(f'prefix: {prefix} | section: {section} | cutter: {cutter}')
```
