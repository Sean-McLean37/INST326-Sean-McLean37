#### 26) Merge two shelves. Given `shelf_a` & `shelf_b`, create `merged` that alternates items from each until one runs out, then appends the rest.

**Solution:**

````python
shelf_a = ['1', '3', '5', '7', '9']
shelf_b = ['2', '4', '6', '8', '10', '11', '12', '13']
merged = []

if len(shelf_a) < len(shelf_b):
    end = len(shelf_a)
    for i in range(end):
        merged.append(shelf_a[i])
        merged.append(shelf_b[i])
    merged += shelf_b[end:]
    
elif len(shelf_b) < len(shelf_a):
    end = len(shelf_b)
    for i in range(end):
        merged.append(shelf_a[i])
        merged.append(shelf_b[i])
    merged += shelf_a[end:]
    
else:
    for i in range(len(shelf_a)):
        merged.append(shelf_a[i])
        merged.append(shelf_b[i])

print(merged)
````
